#### ***7.2 Генеральная и выборочная совокупность***

- *Генеральная совокупность*  - это СВ $X(\omega)$, заданная на пространстве элементарных событий $\Omega$ с выделенным в нем классом подмножеств событий $S$, для которых указаны их вероятности;
- *выборка* - это последовательность $X_1, X_2,..., X_n$ независимых одинаково распределенных СВ, распределение каждой из которых совпадает с распределением генеральной СВ;
- число объектов (наблюдений) в совокупности, генеральной и выборочной, называется ее *объемом*; обозначается соответственно $N$ или $n$;
- конкретные значения выборки, полученные в результате наблюдений (испытаний), называют *реализацией выборки* и обозначают строчными буквами $x_1, x_2,..., x_n$;
- метод статистического исследования, состоящий в том, что на основе изучения выборочной совокупности делается залкючение о всей генеральной совокупности, называется *выборочным*;
- для получения хороших оценок характеристик генеральной свовокпности, необходимо, чтобы выборка была *репрезентативной*, когда согласно закону больших чисел необходимо соблюдение случайности отбора, т.е. все объекты генеральной совокупности должны иметь равные вероятности попасть вы выборку;
- различают выборки с возвращением (повторные) и без возвращение (бесповторные);
- способы отбора: простой, типический, механический и серийный.
